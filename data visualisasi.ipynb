```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# cell diatas ini digunakan untuk mengimport library yang diperlukan untuk mevisualisasikan dan membaca data

baca =  pd.read_csv('crypto_historical_365days.csv')
baca.head()
# cell di atas ini untuk menunjukkan output dari apa yang mereka baca 5 data teratas agar dapat memahami struktur awal data
```

```python
# bisa juga secara keseluruhan dengan koding print(baca)
print(baca)
```

```python
# kita juga dapat melihat jumlah data, tipe data setiap kolom, serta apakah ada nilai kosong
baca.info()
```

```python
# ini juga dapat meringkas/describe data statistik
baca.describe()
```

```python
# kita juga dapat memvisualisai dalam bentuk histogram dengan library matplotlib.pyplot atau plt
baca['coin_id'].hist()
plt.show
```

```python
# disini saya juga coba pakai scatter plot untuk menentukan suatu hubungan antara dua variabel menggunakan library seaborn tadi
sns.scatterplot(x = 'coin_id', y = 'price', data = baca)
plt.show
```
